```python
import pandas as pd
df_test = pd.read_csv('/data/ephemeral/home/AI_Portfolio/Python/data/test.csv')
nullcnt = df_test.isnull().sum()
notnullcnt = df_test.notnull().sum()
full_number = pd.concat([nullcnt + notnullcnt], axis=0)
result_df = pd.DataFrame({
    'null_count': nullcnt,
    'notnull_count': notnullcnt,
    'full_number' : full_number })
display(result_df)
```
