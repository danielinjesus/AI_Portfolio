```python
import pandas as pd
df_train = pd.read_csv('C:\\Daun\\train.csv')
nullcnt = df_train.isnull().sum()
notnullcnt = df_train.notnull().sum()
full_number = pd.concat([nullcnt + notnullcnt], axis=0)
result_df = pd.DataFrame({
    'null_count': nullcnt,
    'notnull_count': notnullcnt,
    'full_number' : full_number })
display(result_df)
```

```python
# df_train['도로명'].value_counts()
df_add = df_train[df_train['도로명'].apply(len) < 3]
display(df_add, len(df_add))
```

```python
# display(df[df['번지'].isnull()])
df_address_needed = df_train[df_train['좌표X'].isnull()][['도로명']]
df_address_needed_index = df_address_needed.index
display(df_address_needed)
display(df_address_needed_index)
```

```python
# 앞서 예측한 예측값들을 저장합니다.
df_address_needed.to_csv('null2-x.csv')
```

```python
%pip install geopy
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="myGeocoder")

# 위도와 경도 열을 미리 생성
df_address_needed['위도'] = None
df_address_needed['경도'] = None

#데이터의 좌표주소 열을 이용해 지오코딩
for i, loc in enumerate(df_address_needed['도로명']):
    coord = geolocator.geocode(loc)[-1]
    df_address_needed.loc[i, '위도'] = coord[0]
    df_address_needed.loc[i, '경도'] = coord[1]
print(df_address_needed)
```
